```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./data/data_lol.csv', index_col=0)

# Columnas para analizar (Agregar mas si es necesario)
blueGoldDiff = df['blueGoldDiff']
blueExperienceDiff = df['blueExperienceDiff']
blueKills = df['blueKills']
blueDeaths = df['blueDeaths']
blueAssists = df['blueAssists']
blueEliteMonster = df['blueEliteMonsters']
blueTowersDestroyed = df['blueTowersDestroyed']
blueAvgLevel = df['blueAvgLevel']
```

```python
# Histogramas
plt.figure(figsize=(20, 20))

plt.subplot(331)
plt.hist(blueGoldDiff, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Diferencia de oro del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueGoldDiff')

plt.subplot(332)
plt.hist(blueExperienceDiff, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Diferencia de experiencia del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueExperienceDiff')

plt.subplot(333)
plt.hist(blueKills, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Asesinatos del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueKills')

plt.subplot(334)
plt.hist(blueDeaths, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Muertes del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueDeaths')

plt.subplot(335)
plt.hist(blueAssists, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Asistencias del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueAssists')

plt.subplot(336)
plt.hist(blueEliteMonster, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Monstruos de élite del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueEliteMonster')

plt.subplot(337)
plt.hist(blueTowersDestroyed, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Torres destruidas del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueTowersDestroyed')

plt.subplot(338)
plt.hist(blueAvgLevel, bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Nivel promedio del equipo azul')
plt.ylabel('Partidas jugadas')
plt.title('blueAvgLevel')

plt.show()
```
