```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./data/data_lol.csv', index_col=0)

# Columnas para analizar (Agregar mas si es necesario)
blueGoldDiff = df['blueGoldDiff']
blueExperienceDiff = df['blueExperienceDiff']
blueKills = df['blueKills']
blueDeaths = df['blueDeaths']
blueAssists = df['blueAssists']
blueEliteMonster = df['blueEliteMonsters']
blueTowersDestroyed = df['blueTowersDestroyed']
blueAvgLevel = df['blueAvgLevel']
blueWins = df['blueWins']

wins = blueWins[blueWins == 1]
losses = blueWins[blueWins == 0]
```

```python
columnas = [blueGoldDiff, blueExperienceDiff, blueKills, blueDeaths, blueAssists, blueEliteMonster, blueTowersDestroyed, blueAvgLevel]
tot_col = len(columnas)

med_col_victorys = []
med_col_losses = []
for i in range(tot_col):
    med_col_victorys.append(np.mean(columnas[i][wins.index]))
    med_col_losses.append(np.mean(columnas[i][losses.index]))
```

```python
# Grafico de barras de las victorias y derrotas
plt.figure(figsize=(30, 10))


# Victorias
plt.subplot(1, 2, 2)
plt.grid(True, alpha=0.2, linestyle='--', linewidth=1, color='black')

nombre_col_wins = []
for i in range(tot_col):
    nombre_col_wins.append(f'{columnas[i].name} - Wins')
    
    plt.text(i, med_col_victorys[i], med_col_victorys[i], ha='center', va='bottom')
    plt.bar(columnas[i].name, med_col_victorys[i])
    plt.legend(nombre_col_wins, loc='upper right')
    
plt.title('Media de los datos para las victorias del equipo azul', 
          fontsize=20, 
          fontweight='bold',
          color='blue',
          alpha=0.8,
          loc='center',)
plt.xlabel('Columnas', fontsize=15, fontweight='bold')
plt.ylabel('Media total para las victorias', fontsize=15, fontweight='bold')
plt.xticks(rotation=45)

# Derrotas
plt.subplot(1, 2, 1)
plt.grid(True, alpha=0.2, linestyle='--', linewidth=1, color='black')

nombre_col_losses = []
for i in range(tot_col):
    nombre_col_losses.append(f'{columnas[i].name} - Losses')
    
    plt.text(i, med_col_losses[i], med_col_losses[i], ha='center', va='bottom')
    plt.bar(columnas[i].name, med_col_losses[i])
    plt.legend(nombre_col_losses, loc='lower right')
    
plt.title('Media de los datos para las derrotas del equipo azul',
            fontsize=20, 
            fontweight='bold',
            color='red',
            alpha=0.8,
            loc='center',)
plt.xlabel('Columnas', fontsize=15, fontweight='bold')
plt.ylabel('Media total para las derrotas', fontsize=15, fontweight='bold')
plt.xticks(rotation=45)

plt.show()
```
