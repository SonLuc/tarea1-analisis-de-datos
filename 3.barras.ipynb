```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from modules import descriptivo as desc

df = pd.read_csv('./data/data_lol.csv', index_col=0)

# Columnas para analizar (Agregar mas si es necesario)
blueGoldDiff = df['blueGoldDiff']
blueExperienceDiff = df['blueExperienceDiff']
blueKills = df['blueKills']
blueDeaths = df['blueDeaths']
blueAssists = df['blueAssists']
blueEliteMonster = df['blueEliteMonsters']
blueTowersDestroyed = df['blueTowersDestroyed']
blueAvgLevel = df['blueAvgLevel']
blueWins = df['blueWins']

wins = blueWins[blueWins == 1]
losses = blueWins[blueWins == 0]
```

```python
columnas = [blueGoldDiff, blueExperienceDiff, blueKills, blueDeaths, blueAssists, blueEliteMonster, blueTowersDestroyed, blueAvgLevel]
tot_col = len(columnas)
```

```python
medias = desc.media(columnas, tot_col)

for i in range(tot_col):
    print(f'Media o promedio de "{columnas[i].name}": {medias[i]}')
```

```python
# Grafico de barras de las victorias y derrotas
plt.figure(figsize=(25, 10))

plt.subplot(1, 2, 1)
plt.bar(['Victorias', 'Derrotas'], [len(wins), len(losses)], color=['blue', 'red'], alpha=0.5)
plt.title('Victorias y derrotas')
plt.xlabel('Resultado')
plt.ylabel('Cantidad de partidas')

# Grafico de barras de las medias
plt.subplot(1, 2, 2)
for i in range(tot_col):
    plt.bar(columnas[i].name, medias[i])
    
plt.title('Medias')
plt.xlabel('Columnas')
plt.ylabel('Media total')
plt.xticks(rotation=45)
plt.show()
```
