```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/data_lol.csv', index_col=0)


blueWins = df['blueWins']
wins = blueWins[blueWins == 1]

# Tomar todas las columnas filtradas por las victorias
blueGoldDiff = df['blueGoldDiff'][wins.index]
blueExperienceDiff = df['blueExperienceDiff'][wins.index]
blueKills = df['blueKills'][wins.index]
blueDeaths = df['blueDeaths'][wins.index]
blueAssists = df['blueAssists'][wins.index]
blueEliteMonster = df['blueEliteMonsters'][wins.index]
blueTowersDestroyed = df['blueTowersDestroyed'][wins.index]
blueAvgLevel = df['blueAvgLevel'][wins.index]
```

```python
columnas = [blueGoldDiff, blueExperienceDiff, blueKills, blueDeaths, blueAssists, blueEliteMonster, blueTowersDestroyed, blueAvgLevel]
tot_col = len(columnas)
```

```python
plt.figure(figsize=(45, 30), dpi=100)


columnas_title = ['diferencia de oro', 'diferencia de experiencia', 'cantidad de asesinatos', 'muertes', 'asistencias', 'cantidad de monstruos de lite', 'torres destruidas', 'nivel promedio']

for i in range(tot_col):
    plt.subplot(4, 5, i+1)
    plt.grid(True, alpha=0.5, linestyle='--', linewidth=1)
    plt.title(f'Boxplot para la {columnas_title[i]} en las victorias del equipo azul', fontsize=15, fontweight='bold')
    plt.boxplot(columnas[i], 
                labels=[columnas[i].name], 
                vert=True, 
                showmeans=True,
                patch_artist=True, 
                boxprops=dict(facecolor='blue', color='blue', alpha=0.5),
                flierprops=dict(marker='o', markerfacecolor='blue', markersize=5, alpha=0.5),
                meanprops=dict(marker='x', markerfacecolor='white', markeredgecolor='blue', markersize=7),
                medianprops=dict(color='orange', linewidth=2),
            )
    plt.xlabel(columnas_title[i].capitalize(), fontsize=10, fontweight='bold')
    plt.ylabel('Cantidad', fontsize=10, fontweight='bold')
    plt.tight_layout()

plt.show()
```
